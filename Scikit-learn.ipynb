# Overview of Scikit-Learn

Scikit-learn is a machine learning package as part of the Python programming language. It is a highly efficient tool that can be used for predictive data analysis.  The application of Scikit-learn can assist the user in developing a deeper understanding of data sets with regard to machine learning algorithms.  Scikit-learn is built on NumPy, SciPy and matplotlib [1]. 

Scikit-learn provides supervised and unsupervised learning algorithms to the user[3].

* **Supervised Machine Learning:**
The majority of practical machine learning uses supervised learning. Supervised learning is where you have input variables (x) and an output variable (Y) and you use an algorithm to learn the mapping function from the input to the output. Y = f(X). The goal is to approximate the mapping function so well that when you have new input data (x) that you can predict the output variables (Y) for that data. It is called *supervised learning* because the process of an algorithm learning from the training dataset can be thought of as a teacher supervising the learning process. We know the correct answers, the algorithm iteratively makes predictions on the training data and is corrected by the teacher. Learning stops when the algorithm achieves an acceptable level of performance [2].


* **Unsupervised Machine Learning:**  Unsupervised Learning is a machine learning technique in which the users do not need to supervise the model. Instead, it allows the model to work on its own to discover patterns and information that was previously undetected. It mainly deals with the unlabelled data. Unsupervised Learning Algorithms allow users to perform more complex processing tasks compared to supervised learning. Although, unsupervised learning can be more unpredictable compared with other natural learning methods. Unsupervised learning algorithms include clustering, anomaly detection, neural networks, etc. [4].



The software is opensource and commercially usable making it highly useful for all businesses.  Scikit-learn has six main areas in which it can be applied [1]:

# First Algorithm- Nearest Neighbors



K-Nearest Neighbors operates by checking the distance from some test example to the known values of some training example. The group of data points/class that would give the smallest distance between the training points and the testing point is the class that is selected [18]. 

KNN can be used for both classification and regression predictive problems. However, it is more widely used in classification problems in the industry. To evaluate any technique we generally look at 3 important aspects [20]:

1. Ease to interpret output

2. Calculation time

3. Predictive Power




**How does the KNN algorithm work?:**
    
If we want to find the classification of the green circle in the diagram below KNN algorithm will take the nearest 3 data points (this figure can be adjusted accordingly) and classify the green circle as either a blue square or a red triangle.

<br/>

```python
# split data set into attributes and labels 
X = df.iloc[:, :-1].values
y = df.iloc[:, 4].values
```

```python
# create train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
# scale features to allow for easier comparison
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# fit classifier to data set (K value of 5 used)
classifier = KNeighborsClassifier(n_neighbors=5)
```

```python
# fit the classifier to the dataclassifier.
classifier.fit(X_train, y_train)
```

After the classifier model has been trained on the training data, it can make predictions on the testing data. This is easily done by calling the predict command on the classifier and providing it with the parameters it needs to make predictions about, which are the features in your testing dataset [17].

```python
# make predictions on test data
y_prediction = classifier.predict(X_test)
```

<br>

<center>Classification Report<center/>


The classification report is a Scikit-Learn built in metric created especially for classification problems. Using the classification report can give you a quick intuition of how your model is performing. **Recall** pits the number of examples your model labeled as Class A (some given class) against the total number of examples of Class A, and this is represented in the report. The report also returns prediction and f1-score. **Precision** is the percentage of examples your model labeled as Class A which actually belonged to Class A (true positives against false positives), and **f1-score** is an average of precision and recall [18].

```python
# use classification Report to view and evaluate the data
print("-------------------------")
print("Classification Results:")
print("-------------------------")
print(classification_report(y_test, y_prediction))
```

<center>Confusion Matrix<center/>
    
        
A confusion matrix is a table or chart, representing the accuracy of a model with regards to two or more classes. The predictions of the model will be on the X-axis while the outcomes/accuracy are located on the y-axis. The cells are filled with the number of predictions the model makes. Correct predictions can be found on a diagonal line moving from the top left to the bottom right. You can read more about interpreting a confusion matrix [18].


```python
# use confusion matrix to view and evaluate the data
print("-------------------------")
print("Confusion Martix Results:")
print("-------------------------")
print(confusion_matrix(y_test, y_prediction))
```

<br/>


```python
# compare error rate with the K-value between 1-40
# loop 1 to 40 and predict mean error

error=[]

for i in range(1, 40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error.append(np.mean(pred_i != y_test))
```

```python
# plot the error values with K-values.
plt.figure(figsize=(16, 8))
plt.plot(range(1, 40), error, color='blue', linestyle='dashed', marker='d',
         markerfacecolor='red', markersize=5)
plt.title('Error Rate K-Value')
plt.xlabel('K-Value')
plt.ylabel('Mean Error')
```

The output demonstrates that the mean error is close to zero when K-values (n=5) are between 6 to 18 along the X-axis.

It halts creating and optimizing clusters when either [19]:
* The centroids have stabilized  there is no change in their values because the clustering has been successful or;

```python
X_data = data.drop(['species', 'species_id'], axis = 1) 
```

```python
# creating K-Means where  n=6
kmeans = KMeans(n_clusters=6)

# fit and predict
y = kmeans.fit_predict(X_data)
print(y)
```

```python
# measures how well a dataset was clustered
kmeans.inertia_
```

Selecting the right K-Means number is critical for a good application of the algorithm.  A number that is far too high can show artificial boundaries within your data set which increases the cost of computation.  On the other hand, if the number is too low then it may create disjointed data into groups of clusters. [8] A popular methodology to address some of these issues uses a method called the Elbow Method  This method essentially runs K K-means clustering on the dataset for a wide variety of values, for instance say 1-10.  For each value 1 to 10 an average cluster is determined. A good model is one with low inertia AND a low number of clusters. [9] 

```python
# Elbow method 
clus_num=[]
for clusters in range(1, 10):
    kmeans = KMeans(n_clusters = clusters).fit(X_data)
    kmeans.fit(X_data)
    clus_num.append(kmeans.inertia_)

plt.plot(range(1, 10), clus_num)
plt.title('Elbow Method')
plt.xlabel('Cluster Number K_values')
plt.ylabel('Inertia')
plt.show()
```

# Thrid Algorithm - Logistic Regression 

Classification is essential, for machine learning., according to some estimates the majority of problems in Data Science is related to how to classify the problem.  The logistics regression is a very common regression method for helping to solve binary classification problems.  For example, it can be used for Diabetes  detection problems. Moreover, Logistic regression is easy to use and is widely used.  The main difference between logistics regression and linear regression is the fact that linear regression gives you a continuous output, but logistics regression gives you a constant output. For instance, a continuous output might be the price of a Bitcoin where a discrete output is deterministic, you either have cancer or not

There are three main types of Logistics Regression:

	**Binary Logistic Regression:** Binary logistic regression is used to predict the odds of being a case based on the values of the independent variables (predictors) [10].


	**Multinomial Logistic Regression:**  Multinomial logistic regression (often just called 'multinomial regression') is used to predict a nominal dependent variable given one or more independent variables. It is sometimes considered an extension of binomial logistic regression to allow for a dependent variable with more than two categories [11].

	**Ordinal Logistic Regression:** Ordinal logistic regression is an extension of simple logistic regression model.  in ordinal logistic regression the dependent variable is ordinal i.e. there is an explicit ordering in the categories. For instance customer data was gathered on a rating equal to X or less than X for a product [12].

To showcase the logistics regression the "Pima Indians Diabetes Database" will be downloaded from Kagggle.com and uploaded to Github [13]

<center> Confusion Matrix Evaluation Metrics <center>


Accuracy is probably one of the easiest evaluation metrics to understand but is also often the least relevant. Accuracy measures the total number of correct predictions a classification model makes, including both true positives and true negatives. You can use the formula below to calculate [14]:






```python
metrics.accuracy_score(y_test, y_pred)
```


Precision measures the number of true positives out of all predicted positives. Conservative models often have very high precision scores, because there is a high threshold for predicting a positive. It is calculated as [15]:









```python
metrics.precision_score(y_test, y_pred)
```

Recall measures the number of true positives out of all actual total positives (true positives plus false negatives). You may also hear recall referred to as the true positive rate. This can be calculated as [16]:







```python
metrics.recall_score(y_test, y_pred)
```

```python
print("--------------------------------------")
print("Evaluation Metrics of Log Reg Model")
print("--------------------------------------")
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
print("---------------------------------------")
```

A classification rate of 80% is considered as good accuracy.  The precision is all about how precise the model is, in order words you could say how often is your model making right predictions. The above model suggests that right predictions will be made approximately 76% of the time.  The Recall: finds if there are persons who have diabetes within the test set how often the log regression can identify them.  In this instance 58% of the time they will be identified [17].

<center> The Receiver Operating Characteristic (ROC) <center>
    <br>
    
The ROC is a curve plot which shows the positive rate against the false positive rate. 
It is a plot of the false positive rate (x-axis) versus the true positive rate (y-axis) for a number of different candidate threshold values between 0.0 and 1.0. Put another way, it plots the false alarm rate versus the hit rate.

    
    

```python
# ROC model using the train/test data that we split earlier
# predict possibilities
y_pred_proba = logreg.predict_proba(X_test)[::,1]

# calculate curves and scores
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)

# plot graph
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

The  Area Under the roc Curve (AUC) score in this case is 0.86. A score of 1 would determine a perfect score for the classifier while a score of 0.5 would represent a worthless classifier!
